## removing outliers
#### for all columns in one go rather than sequentially



```python
def remove_outliers(df, col_name):
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    if abs(q1 - q3) > 0.1 and df[col_name].nunique() > 2:  # basically q1 and q3 should not be equal
        iqr = q3-q1
        lower_lim = q1 - 1.5 * iqr
        upper_lim = q3 + 1.5 * iqr
        df = df.loc[(df[col_name]>=lower_lim) & (df[col_name] <= upper_lim)]
    return df
```

```python
model1_ = sm.GLM(y_train_sm, sm.add_constant(X_train_sm), family=sm.families.Binomial())
model1 = model1_.fit()
model1.summary()
```

```python
def print_metrics(y_act, y_pred=None, type_='Train', y_pred_proba=None):
    if y_pred is not None:
        print(type_, 'confusion_matrix', confusion_matrix(y_act, y_pred))
        print(type_, 'accuracy_score', accuracy_score(y_act, y_pred))
        print(type_, 'precision score', precision_score(y_act, y_pred))
        print(type_, 'recall', recall_score(y_act, y_pred))
    if y_pred_proba is not None:
        print(type_, 'ROC AUC score', roc_auc_score(y_act, y_pred_proba))
```

```python
y_train_sm_pred_proba = model1.predict(sm.add_constant(X_train_sm))
```

```python
fpr, tpr, thresholds = roc_curve(y_train_sm, y_train_sm_pred_proba)
print(fpr)
print(tpr)
print(thresholds)
```

```python
len(thresholds)
```

```python
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel("TPR")
```

```python
# get the optimal thresholds
def get_optimal_thresholds(fpr, tpr, thresholds):
    g_mean = (tpr*(1-fpr))**0.5
    index = np.argmax(g_mean)
    thresh_opt = round(thresholds[index], 3)
    fpr_opt = round(fpr[index], 3)
    tpr_opt = round(tpr[index], 3)
    return thresh_opt, fpr_opt, tpr_opt
```

```python
thresh_opt, fpr_opt, tpr_opt = get_optimal_thresholds(fpr, tpr, thresholds)
```

```python
y_train_sm_pred = (y_train_sm_pred_proba >= thresh_opt).astype('int')
```

```python
y_train_sm_pred[:3]
```
